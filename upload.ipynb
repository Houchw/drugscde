```python
#!/usr/bin/env python
# -*- coding: utf-8 -*-
import requests
import pandas as pd
import numpy as np
import json
import re
```

```python
with open("drugsInfo_1.jl", "rb") as f:
    data = [json.loads(line.strip()) for line in f.readlines()]

df = pd.DataFrame(data)
df.rec_date = pd.to_datetime(df.rec_date)
df = df.drop_duplicates()
df.head()
```

```python
df.rec_date = df.rec_date.fillna('1900-01-01')
```

```python
def companyForm(s):
    pattern = re.compile(u'( +[a-zA-Z0-9]*[\u4e00-\u9fa5]+\S*[$\u4e00-\u9fa5])')
    company = re.split(pattern, s)
    #print(company)
    company = [i.strip() for i in company if i != ""]
    company = list(set(company))
    company.sort()
    company = "|".join(company)
    return company
```

```python
companyForm("河南省新乡联谊制药厂 河南省新乡联谊制药厂")
```

```python
df.corp_name = df.corp_name.map(companyForm)
df.corp_name = df.corp_name.replace("", u"我确定不知道是啥公司")
```

```python
corp_name = df.corp_name.drop_duplicates().values
```

```python
df.app_type = df.app_type.replace("", "None")
df.drug_name = df.drug_name.replace("", u"药物名称")
df.drug_type = df.drug_type.replace("", u"药物类型")
df.inv_type = df.inv_type.replace("", u'注册分类')
```
